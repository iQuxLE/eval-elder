```python
import numpy as np
from typing import List, Dict

# Example function to simulate the HPO clustering object
def get_organ_system(hpo_term):
    cluster_map = {
        'HP:001': 'organ1',
        'HP:002': 'organ1',
        'HP:003': None,  # This term does not belong to any cluster
        'HP:004': 'organ2',
        'HP:005': 'organ2'
    }
    return cluster_map.get(hpo_term, None)

# Example function to simulate the embeddings dictionary
def get_hp_embedding(hpo_term):
    embedding_map = {
        'HP:001': np.array([1, 2, 3]),
        'HP:002': np.array([4, 5, 6]),
        'HP:003': np.array([7, 8, 9]),
        'HP:004': np.array([10,11,12]),
        'HP:005': np.array([13, 14, 15])
    }
    return embedding_map.get(hpo_term, -1 * np.ones(3))

# Mocked data
hp_embeddings = {
    'HP:001': np.array([1, 2, 3]),
    'HP:002': np.array([4, 5, 6]),
    'HP:003': np.array([7, 8, 9]),
    'HP:004': np.array([10,11,12]),
    'HP:005': np.array([13, 14, 15])
}

all_organ_systems = ['organ1', 'organ2', 'organ3']
disease_to_hps = {
    'Disease1': ['HP:001', 'HP:003', "HP:005", "HP:004"],
    'Disease2': ['HP:002', 'HP:003', "HP:001"],
}

# Function to compute organ embeddings
def compute_organ_embeddings(hpo_terms: List[str] = None):
    embedding_size = 3
    organ_system_embeddings = {organ: [] for organ in all_organ_systems}
    terms_without_cluster = []

    for hpo_term in hpo_terms:
        organ_system = get_organ_system(hpo_term)
        if organ_system is None:
            terms_without_cluster.append(hpo_term)
            continue

        embedding = get_hp_embedding(hpo_term)
        organ_system_embeddings[organ_system].append(embedding)

    concatenated_embeddings = []
    for organ_system in all_organ_systems:
        if organ_system_embeddings[organ_system]:
            average_embedding = np.mean(organ_system_embeddings[organ_system], axis=0)
        else:
            average_embedding = -1 * np.ones(embedding_size)
        concatenated_embeddings.append(average_embedding)

    final_embedding = np.concatenate(concatenated_embeddings)
    return final_embedding, terms_without_cluster

# Testing the function with mock data
hpo_terms_for_disease1 = disease_to_hps['Disease1']
embedding_disease1, terms_without_cluster_disease1 = compute_organ_embeddings(hpo_terms_for_disease1)

embedding_disease1

```

```python
def create_hpo_id_to_embedding(collection: Collection) -> Dict:
    """
    Create a dictionary mapping HPO IDs to embeddings.

    :param collection: The collection to process
    :return: A dictionary mapping HPO IDs to a dictionary of their label and embeddings.
    """
    hpo_id_to_data = {}
    results = collection.get(include=["metadatas", "embeddings"])
    for metadata, embedding in zip(results.get("metadatas", []), results.get("embeddings", []), strict=False):
        metadata_json = json.loads(metadata["_json"])
        hpo_id = metadata_json.get("original_id")
        if hpo_id:
            hpo_id_to_data[hpo_id] = {"embeddings": embedding}  # #{'HP:0005872': [1,2,3, ...]}
    return hpo_id_to_data
```

```python
def create_disease_to_hps_dict(collection: Collection) -> Dict:
    """
    Creates a dictionary mapping diseases (OMIM IDs) to their associated HPO IDs.

    :param collection: The collection to process
    :return: Dictionary with diseases as keys and lists of corresponding HPO IDs as values.
    """
    disease_to_hps_dict = {}
    results = collection.get(include=["metadatas"])
    for item in results.get("metadatas"):
        metadata_json = json.loads(item["_json"])
        disease = metadata_json.get("disease")
        phenotype = metadata_json.get("phenotype")
        # label = metadata_json.get("disease_label")
        if disease and phenotype:
            if disease not in disease_to_hps_dict:
                disease_to_hps_dict[disease] = [phenotype] #put the label
            else:
                disease_to_hps_dict[disease].append(phenotype) # put the label
    print(len(disease_to_hps_dict)) # used to be 2647, now 3233
    return disease_to_hps_dict
```

```python

# hp_embeddings_dict = runner.data_processor.hp_embeddings
# len(hp_embeddings_dict) # should be 29k
# dis_hp_from_12k_dict = runner.db_manager.create_collection("dis_hp_from_12k_dict")
import time
def process_data(dis_hp_from_12k_dict) -> Collection:

    batch_size = 100
    batch = []
    embedding_calc_time = 0
    upsert_time = 0

    for disease, hps in dictionary_omim.items():
        start = time.time()
        average_embedding = calculate_average_embedding(hps, hp_embeddings_dict)
        embedding_calc_time += time.time() - start
        batch.append((disease, average_embedding.tolist()))

        if len(batch) >= batch_size:
            start = time.time()
            upsert_batch(batch, dis_hp_from_12k_dict)
            upsert_time += time.time() - start
            batch = []

    if batch:
        start = time.time()
        upsert_batch(batch, dis_hp_from_12k_dict)
        upsert_time += time.time() - start

    print(f"Total time for embedding calculations: {embedding_calc_time}s")
    print(f"Total time for upsert operations: {upsert_time}s")

    return dis_hp_from_12k_dict

def upsert_batch(batch, dis_hp_from_12k_dict):
    ids = [item[0] for item in batch]
    embeddings = [item[1] for item in batch]
    metadatas = [{"type": "disease"}] * len(batch)
    dis_hp_from_12k_dict.upsert(ids=ids, embeddings=embeddings, metadatas=metadatas)
    
def calculate_average_embedding(hps: list, embeddings_dict: Dict) -> np.ndarray:
    embeddings = [embeddings_dict[hp_id]["embeddings"] for hp_id in hps if hp_id in embeddings_dict]
    return np.mean(embeddings, axis=0) if embeddings else []
```
